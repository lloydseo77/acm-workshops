```python
# Using coordinates
import osmnx
import geopandas

dtsd_coord = (32.7174202, -117.1627720)
graph = osmnx.graph_from_point(dtsd_coord, 1000)
```

```python
figure, ax = osmnx.plot_graph(graph)
```

```python
nodes, edges = osmnx.graph_to_gdfs(graph)
nodes.head()
```

```python
edges.head()
```

```python
features = osmnx.features_from_point(dtsd_coord, tags={'building': True}, dist=1000)
```

```python
features
```

```python
features.plot()
```

```python
buildings = osmnx.features_from_point(
    dtsd_coord,
    {"building": True},
)
```

```python
len(features)
```

```python
features.columns
```

```python
restaurants = osmnx.features_from_point(
    dtsd_coord,
    {
        "amenity": "restaurant"
    }
)
len(restaurants)
```

```python
restaurants.columns.values
```

```python
# Select some useful cols and print
interesting_columns = [
    "name",
    "opening_hours",
    "reservation",
    "cuisine",
    "addr:street",
    "website"
]

# Print only selected cols
restaurants = restaurants.sort_values(by="name")
restaurants[interesting_columns].head(10)
```

```python
parks = osmnx.features_from_point(
    dtsd_coord,
    {
        "leisure": "park",
        "landuse": "grass",
    },
)
```

```python
parks.head()
```

```python
parks.plot(color="green")
```

```python
import matplotlib
figure, ax = matplotlib.pyplot.subplots(figsize=(12,8))

# Plot the footprint
features.plot(ax=ax, facecolor="black")

# Plot parks
parks.plot(ax=ax, facecolor="green")

# Plot street edges
edges.plot(ax=ax, linewidth=1, edgecolor="dimgray")

# Plot buildings
buildings.plot(ax=ax, facecolor="silver", alpha=0.7)

# Plot restaurants
restaurants.plot(ax=ax, color="yellow", alpha=0.7, markersize=10)
```
