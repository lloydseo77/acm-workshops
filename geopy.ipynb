```python
import pathlib
NOTEBOOK_PATH = pathlib.Path().resolve()
DATA_DIRECTORY = NOTEBOOK_PATH / "drive" / "MyDrive" / "Colab Notebooks"
```

```python
import pandas
addresses = pandas.read_csv(DATA_DIRECTORY / "uc_addresses.csv")
addresses.head(10)
```

```python
import geopandas

geocoded_addresses = geopandas.tools.geocode(
    addresses["addr"],
    provider="nominatim",
    user_agent="lloydseo",
    timeout=10
)
geocoded_addresses.head(10)
```

```python
geocoded_addresses_with_id = geocoded_addresses.join(addresses)
geocoded_addresses_with_id
```

```python
# Save geocoded addresses to a GeoPackage file
geocoded_addresses.to_file(DATA_DIRECTORY / "uc_addresses.gpkg")
```
