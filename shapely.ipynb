```python
# Import `shapely.geometry.Point` class
from shapely.geometry import Point

# Create `Point` objects:
point1 = Point(2.2, 4.2)
point2 = Point(7.2, -25.1)
point3 = Point(9.26, -2.456)
point4_3D = Point(9.26, -2.456, 0.57)
point1
```

```python
# Print coordinates -- this is WKT formatting
# Z indicates coordinate is in 3D

print(point1)
print(point4_3D)
```

```python
# Print list of coordinate tuples
list(point1.coords)
```

```python
# Points only contain one coordinate tuple,
# so they have direct access to coordinates
x = point1.x
y = point1.y

print(x, y)
```

```python
# Calculate the distance between point1 and point2
dist = point1.distance(point2)
print(f"Distance between the points is {dist:.2f} units")
```

```python
# import the LineString class
from shapely.geometry import LineString

# Create a LineString from our Point objects
line = LineString([point1, point2, point3])
line
```

```python
# LineString can be constructed through coordinate tuples
line2 = LineString([(2.2, 4.2), (7.2, -25.1), (9.26, -2.456)])
```

```python
line == line2
```

```python
# Print coordinate tuples
print(line)
```

```python
# Get coordinate tuples
list(line.coords)
```

```python
# Obtain x and y coordinates
xcoords = list(line.xy[0])
ycoords = list(line.xy[1])

print(xcoords)
print(ycoords)
```

```python
# Get the length of the line
line_length = line.length
print(f"Length of our line: {line_length:.1f} units")
```

```python
# Get the centre point of the line -- average of
print(line.centroid)
```

```python
from shapely.geometry import Polygon

# Create a Polygon from coordinates
polygon1 = Polygon([(2.2, 4.2), (7.2, -25.1), (9.26, -2.456)])
polygon1
```

```python
# Construction using points
polygon2 = Polygon([point1, point2, point3])
```

```python
# LinearRings are similar to LineStrings
from shapely.geometry import LinearRing

# Construction using LinearRings
shell = LinearRing([point1, point2, point3, point1])
polygon3 = Polygon(shell)
shell
```

```python
polygon3
```

```python
polygon1 == polygon2 == polygon3
```

```python
# Print coordinate tuples
print(polygon1)
```

```python
# define the exterior
outer = LinearRing([(-180, 90), (-180, -90), (180, -90), (180, 90)])

# define a hole:
hole = LinearRing([(-170, 80), (-100, -80), (100, -80), (170, 80)])
```

```python
outer
```

```python
hole
```

```python
polygon_without_hole = Polygon(outer)
polygon_without_hole
```

```python
# Note second argument is a list
polygon_with_hole = Polygon(outer, [hole])
polygon_with_hole
```

```python
print(polygon_without_hole)
print(polygon_with_hole)
```

```python
# @title
import matplotlib.pyplot as plt

polygon = Polygon([(0, 0), (1, 0), (1, 1), (0, 1)])

# Create a buffer with a negative distance
contracted_polygon = polygon.buffer(-0.2)

# Plot the original Polygon and the contracted Polygon
plt.plot(*polygon.exterior.xy, label='Original Polygon')
plt.plot(*contracted_polygon.exterior.xy, label='Contracted Polygon')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
